```python
# Importing necessary packages
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn import linear_model
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import cross_val_predict, cross_val_score, train_test_split


# Reading the dataset
data = pd.read_csv("../project-3-house-prices/train.csv")
```

```python
# take a look at the data
data.head()
```

```python
# these are the only columns I will be using for this project. 
columns_to_use = ["LotArea", "Utilities","Neighborhood","BldgType","HouseStyle", "OverallQual","OverallCond",\
                  "YearBuilt","YearRemodAdd","RoofStyle","RoofMatl","GrLivArea","FullBath","HalfBath",\
                  "BedroomAbvGr","KitchenAbvGr","MoSold","YrSold","SalePrice"]
```

```python
# create a dataframe with the only necessary column and name it "df"
df = data[columns_to_use]
```

```python
# creating barplots for variable with object type and exploring their relation with "SalePrice"
for i in range(len(df.columns)):
    if df[df.columns[i]].dtype == object:
        sns.barplot(df["SalePrice"], df[df.columns[i]])
        plt.title("Difference between the mean of SalePrice and String objects")
        plt.xlabel("Mean SalePrice")
        plt.ylabel(str(df.columns[i]))
        plt.show()
```

```python
# creating regplots to see the relationship of all columns with SalePrice
for i in range(len(df.columns)):
    if df[df.columns[i]].dtype == int:
        sns.regplot(df["SalePrice"], df[df.columns[i]])
        plt.title("Difference between SalePrice and other variables")
        plt.xlabel("SalePrice")
        plt.ylabel(str(df.columns[i]))
        plt.show()
```

```python
# Creating dummy variable. 
categories = ['Utilities','Neighborhood',"BldgType", "HouseStyle","RoofStyle", "RoofMatl"]
for category in categories:
   series = df[category]
   dummies = pd.get_dummies(series, prefix=category, drop_first=True)
   df = pd.concat([df, dummies], axis=1)
print df.columns
```

```python
print df.info()
print df.shape
```

```python
# another heat map including dummy variables
correl = df.corr()**2
fig, ax = plt.subplots(figsize=(16,10))
sns.heatmap(correl);
sns.set(style = "white")
plt.show()
```

```python
# choose my x values
x = pd.concat([df.ix[:,"LotArea"],df.ix[:,"OverallQual":"YearRemodAdd"],df.ix[:,"GrLivArea":]], axis = 1)
```

```python
# choose SalePrice as y values
y = df.ix[:,"SalePrice"]
```

```python
df.head()
```

```python
# Doing normalization. 

variables_to_scale = ["LotArea",'YearBuilt', 'YearRemodAdd', 'GrLivArea', 'YrSold']

df.reset_index(inplace=True)
scaler = MinMaxScaler()
df[variables_to_scale] = pd.DataFrame(scaler.fit_transform(df[variables_to_scale]),columns= variables_to_scale)
```

```python
# Split the dataset into training set and test sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.33)
```

```python
# Create a linear regression model 
lm = linear_model.LinearRegression()
```

```python
#fit the model using training sets. 
model = lm.fit(X_train, y_train)
```

```python
# Predict y for x test set. 
Y_pred = model.predict(X_test)
```

```python
# calculate r2
r2 = r2_score(y_true=y_test, y_pred = Y_pred)
r2
```

```python
# find mean square errors
mean_squared_error(y_test, Y_pred)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# create a RidgeCV model 
lm = linear_model.RidgeCV(alphas = (.1,1,10))
```

```python
# fit model 
model = lm.fit(X_train, y_train)
```

```python
# predict Y 
Y_Pred_R = model.predict(X_test)
```

```python
# find r^2
r2 = r2_score(y_true = y_test, y_pred = Y_Pred_R)
```

```python
# show r^2
r2
```

```python
# find the mse
mean_squared_error(y_test, Y_Pred_R)
```

```python
# model cooef's
model.coef_
```

```python
# model intercept
model.intercept_
```

```python
# creating a Lasso model
lm = linear_model.LassoCV()
```

```python
#fitt model !
model = lm.fit(X_train, y_train)
```

```python
# predict y for based on the model for X_test
Y_pred_L = model.predict(X_test)
```

```python
Y_pred_L
```

```python
# find r^2 
r2 = r2_score(y_true = y_test, y_pred = Y_pred_L)
```

```python
# really really good score 
r2
```

```python
# So far, better than all other models. 
mean_squared_error(y_test, Y_pred_L)
```
